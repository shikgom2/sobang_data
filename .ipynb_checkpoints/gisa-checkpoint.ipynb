```python
second = {}
for i in range(len(data)):
    if pd.isna(data['기사'].loc[i]):
        continue
    second[str(data['지점'].loc[i])+"@"+data['지점명'].loc[i]+"@"+data['일시'].loc[i]] = data['기사'].loc[i]
    
for i in second:
    second[i] = second[i].split('{')
    
for i in second:
    for j in range(len(second[i])):
        second[i][j] = second[i][j].split('}')
        
for i in second:
    for j in range(len(second[i])):
        for a in range(len(second[i][j])):
            if second[i][j][a] == '':
                del second[i][j][a]
            elif len(second[i][j]) == 1 and second[i][j][a] == '-':
                del second[i][j][a]
                
for i in second:
    for j in range(len(second[i])):
        for a in range(len(second[i][j])):
            second[i][j][a] = second[i][j][a].strip("."" ")

for i in second:
    for j in range(len(second[i])):
        if len(second[i][j]) <= 1:
            second[i][j+1] = (second[i][j]+second[i][j+1])

third = {}
for i in second:
    third[i] = []
    for j in range(len(second[i])):
        if len(second[i][j]) >= 2:
            third[i].append(second[i][j])
```

```python
기사 = pd.DataFrame(columns=['지점,시간','시간','예보내용','강도'])
for i in third:
    for j in range(len(third[i])):
        if len(third[i][j]) == 2:
            기사.loc[len(기사)] = [i,third[i][j][1],third[i][j][0],None]
        elif len(third[i][j]) == 3:
            기사.loc[len(기사)] = [i,third[i][j][2],third[i][j][0],third[i][j][1][-1]]
```

```python
기사[['지점','지점명','일시']] = 기사['지점,시간'].str.split('@',n=2,expand=True)
기사[['시작시간','끝시간']] = 기사['시간'].str.split('-',n=1,expand=True)
기사 = 기사.reindex(columns=['지점','지점명','일시','시작시간','끝시간','예보내용','강도'])
기사
```
