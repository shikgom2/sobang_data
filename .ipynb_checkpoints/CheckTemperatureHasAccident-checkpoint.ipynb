```python
import pandas as pd
import numpy as np

df = pd.read_csv("datas.csv")

temp_dict = {}
count_dict = {}

rain_dict = {}
wind_dict = {}
humi_dict = {}
snow_dict = {}
giap_dict = {}
iljo_dict = {}
ilsa_dict = {}
cloud_dict = {}
visib_dict = {}

#기온
for index, row in df.iterrows():
    temp = row['기온(°C)']
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in temp_dict:
        temp_dict[str(temp)] = ""
        
    if(temp_dict[str(temp)] == ""):
        temp_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])

    else:
        temp_dict[str(temp)] = str(temp_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#강수량
for index, row in df.iterrows():
    temp = row['강수량']
    if(temp == 'nan'):
        temp = 0
        
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in rain_dict:
        rain_dict[str(temp)] = ""
    
    if(rain_dict[str(temp)] == ""):
        rain_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        rain_dict[str(temp)] = str(rain_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])
        
#풍속
for index, row in df.iterrows():
    temp = row['풍속(m/s)']
    if(temp == " "):
        temp = 0.0
    
    t = float(temp)
    t = round(t,1)
    
    temp = t
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in wind_dict:
        wind_dict[str(temp)] = ""
    
    if(wind_dict[str(temp)] == ""):
        wind_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        wind_dict[str(temp)] = str(wind_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#습도
for index, row in df.iterrows():
    temp = row['습도(%)']
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in humi_dict:
        humi_dict[str(temp)] = ""
    
    if(humi_dict[str(temp)] == ""):
        humi_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        humi_dict[str(temp)] = str(humi_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#적설량
for index, row in df.iterrows():
    temp = row['적설(cm)']
    if(temp == " "):
        temp = 0.0
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in snow_dict:
        snow_dict[str(temp)] = ""
    
    if(snow_dict[str(temp)] == ""):
        snow_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        snow_dict[str(temp)] = str(snow_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#증기압
for index, row in df.iterrows():
    temp = row['증기압(hPa)']
    acdnt = row['acdnt_cause_asort_nm']
    t = float(temp)
    t = round(t,1)
    temp = t
    if str(temp) not in giap_dict:
        giap_dict[str(temp)] = ""
    
    if(giap_dict[str(temp)] == ""):
        giap_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        giap_dict[str(temp)] = str(giap_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#일조량
for index, row in df.iterrows():
    temp = row['일조(hr)']
    if(temp == " " or temp == "0"):
        temp = "0.0"
    if(temp == "1"):
        temp = "1.0"
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in iljo_dict:
        iljo_dict[str(temp)] = ""
    
    if(iljo_dict[str(temp)] == ""):
        iljo_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        iljo_dict[str(temp)] = str(iljo_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#일사량
for index, row in df.iterrows():
    temp = row['일사(MJ/m2)']
    if(temp == ' '):
        temp = '0.00'

    t = float(temp)
    t = round(t,2)
    temp = t
    
    acdnt = row['acdnt_cause_asort_nm']

    if str(temp) not in ilsa_dict:
        ilsa_dict[str(temp)] = ""
    
    if(ilsa_dict[str(temp)] == ""):
        ilsa_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        ilsa_dict[str(temp)] = str(ilsa_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])

#중하층운량
for index, row in df.iterrows():
    temp = row['중하층운량(10분위)']
    if(temp == ' '):
        temp = '0'
    acdnt = row['acdnt_cause_asort_nm']
    
    if str(temp) not in cloud_dict:
        cloud_dict[str(temp)] = ""
    
    if(cloud_dict[str(temp)] == ""):
        cloud_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        cloud_dict[str(temp)] = str(cloud_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])
        
#시정
for index, row in df.iterrows():
    temp = row['시정(10m)']

    acdnt = row['acdnt_cause_asort_nm']
    
    if str(temp) not in visib_dict:
        visib_dict[str(temp)] = ""
    
    if(visib_dict[str(temp)] == ""):
        visib_dict[str(temp)] = str(row['acdnt_cause_asort_nm'])
    else:
        visib_dict[str(temp)] = str(visib_dict[str(temp)]) + "," + str(row['acdnt_cause_asort_nm'])
```

# 1. 기온별 수난사고 구분

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datas.csv")

min_temp = min(df['기온(°C)']) #-2.5
max_temp = max(df['기온(°C)']) #35.7

temperature_df = pd.DataFrame(np.arange(min_temp, max_temp+0.1, 0.1))

temperature_df = temperature_df.round(1)
temperature_df['has_acdnt'] = ""
temperature_df['acdnt'] = ""

pd.set_option('display.max_rows', None) ## 모든 열을 출력한다.

for i in range(len(temperature_df)):
    if str(temperature_df[0][i]) in temp_dict.keys():
        temperature_df['has_acdnt'][i] = "O"
        temperature_df['acdnt'][i] = str(temp_dict.get(str(temperature_df[0][i])))
    else:
        temperature_df['has_acdnt'][i] = "X"
temperature_df
```
