```python
import pandas as pd
import numpy as np

df = pd.read_csv("datas.csv")

temp_dict = {}
count_dict = {}

for index, row in df.iterrows():
    temp = row['기온(°C)']
    accident = row['acdnt_cause_asort_nm']

    if str(temp) not in temp_dict:
        temp_dict[str(temp)] = ""
    
    if(temp_dict[str(temp)] == ""):
        temp_dict[str(temp)] = str(accident)
    else:
        temp_dict[str(temp)] = str(temp_dict[str(temp)]) + "," + str(accident)
temp_dict
```
